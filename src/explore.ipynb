```python
import pandas as pd


df= pd.read_csv('https://raw.githubusercontent.com/4GeeksAcademy/data-preprocessing-project-tutorial/main/AB_NYC_2019.csv')
df
```

```python
df.to_csv('../data/raw/raw_data.scv',index=False)
```

```python
df.info()
```

```python
df.drop("id", axis = 1).duplicated().sum()
```

```python
#eliminar informacion irrelevante
irrelevant_columns=['name','id','host_id','host_name','reviews_per_month','last_review']
df.drop(irrelevant_columns, axis=1, inplace=True)
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig, axis = plt.subplots(2, 3, figsize = (10, 7))

sns.histplot(ax = axis[0, 0], data = df, x = "neighbourhood_group").set(ylabel=None, xticks=[])
sns.histplot(ax=axis[0,1],data= df, x = "neighbourhood" ).set(ylabel = None, xticks=[])
sns.histplot(ax=axis[0,2],data= df, x = "room_type" ).set_xticklabels(axis[0,2].get_xticklabels(),rotation=30, ha="right")
sns.histplot(ax=axis[1,0],data=df,x="availability_365").set(ylabel=None,xticks=[])

plt.tight_layout()

plt.show()

#las variables categoricas son las que dicen object
```
